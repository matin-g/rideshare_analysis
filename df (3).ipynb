```python
trips_with_index = results_df.set_index('trip_id')
trips_with_index
```

```python
f = '%Y-%m-%dT%H:%M:%S.000'
trips_with_start_times = trips_with_index.assign(start_year=trips_with_index.get('trip_start_timestamp').apply(lambda x: datetime.datetime.strptime(x, f).year),
                                                 start_month=trips_with_index.get('trip_start_timestamp').apply(lambda x: datetime.datetime.strptime(x, f).month),
                                                 start_day=trips_with_index.get('trip_start_timestamp').apply(lambda x: datetime.datetime.strptime(x, f).day),
                                                 start_weekday=trips_with_index.get('trip_start_timestamp').apply(lambda x: datetime.datetime.strptime(x, f).weekday()),
                                                 start_hour=trips_with_index.get('trip_start_timestamp').apply(lambda x: datetime.datetime.strptime(x, f).hour)
                                                )
trips_with_start_times
```

```python
g = (trips_with_index.get('trip_end_timestamp').apply(lambda x: datetime.datetime.strptime(x, f)) - trips_with_index.get('trip_start_timestamp').apply(lambda x: datetime.datetime.strptime(x, f))).apply(datetime.timedelta.total_seconds)

trips = trips_with_start_times.assign(duration=g)
trips
```

```python
monthly_trips =trips.groupby(["start_weekday","start_hour"]).count()
monthly_trips.plot(kind='line', y='trip_miles').set_ylabel("number_of_trips")
```

```python
fares = trips.get("fare").apply(float)
tips  = trips.get("tip").apply(float)
proportion = tips/fares
median_tip_proportion = proportion.mean()
median_tip_proportion


tips_scatter = bpd.DataFrame().assign(fares=fares, proportion=proportion)
tips_scatter = tips_scatter[tips_scatter.get("fares") < 60]
tips_scatter = tips_scatter[tips_scatter.get("proportion") < 1]

tips_scatter.plot(kind="scatter", x="fares", y="proportion", figsize=(10, 7), alpha=.1)
```
